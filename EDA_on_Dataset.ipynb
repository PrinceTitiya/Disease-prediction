## <b>Univariate Analysis

```python
print("Total 4 Unique diease mentioned in dataset:",data["Disease"].unique())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Age', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Univariate analysis of gender using a bar plot
frequency_counts_gender = data['Gender'].value_counts()
print("\nFrequency Counts for Gender:")
print(frequency_counts_gender)

plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Gender', palette='pastel')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Univariate analysis of diseases using a bar plot
frequency_counts_disease = data['Disease'].value_counts()
print("\nFrequency Counts for Disease:")
print(frequency_counts_disease)


plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='Disease', palette='Set2', order=data['Disease'].value_counts().index)
plt.title('Distribution of Diseases')
plt.xlabel('Count')
plt.ylabel('Disease')
plt.show()
```

```python
symptom_counts = data.iloc[:, 2:-1].apply(pd.Series.value_counts).sum()

# Plot bar plot for symptom frequency
plt.figure(figsize=(20, 10))
sns.barplot(x=symptom_counts.index, y=symptom_counts.values, color='skyblue')
plt.title('Frequency of Symptoms')
plt.xlabel('Symptom')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

##  Multivariate Analysis:

 <li>how age, gender, and symptoms interact with disease?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Disease', y='Age', hue='Gender', palette='Set2')
plt.title('Age vs. Disease by Gender')
plt.xlabel('Disease')
plt.ylabel('Age')
plt.show()
```

```python
sns.pairplot(data, vars=['Age', 'Fever', 'Cough', 'Headache'], hue='Disease', palette='Set1')
plt.title('Scatter Plot of Age vs. Symptoms')
plt.show()
```

<li>Investigate interactions between symptoms and diseases using a heatmap correlation matrix:</li>

```python
# Select columns for correlation analysis (excluding 'Gender' and 'Disease')
correlation_data = data.drop(columns=['Gender', 'Disease'])

# Compute correlation matrix
correlation_matrix = correlation_data.corr()

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation between Age and Symptoms')
plt.show()
```
